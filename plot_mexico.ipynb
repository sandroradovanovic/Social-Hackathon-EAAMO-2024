### Load Mexico shape file  
Source: Marco Geoestadístico INEGI (2023).

#### URL: https://www.inegi.org.mx/temas/mg/ 

`INEGI data`: This is a comment indicating that the following code deals with INEGI data. INEGI (Instituto Nacional de Estadística y Geografía) is Mexico's National Institute of Statistics and Geography.

```python
# INEGI data
gdf = geopd.read_file("../dataset/Mexico/geometry/2023_1_00_MUN/2023_1_00_MUN.shp")
```

`gdf = geopd.read_file("../dataset/Mexico/geometry/2023_1_00_MUN/2023_1_00_MUN.shp")`: This line reads a shapefile (a common geospatial vector data format) using GeoPandas. The file path points to the location of the shapefile on the disk. The read file is stored in the variable gdf as a GeoDataFrame, which is a data structure similar to a pandas DataFrame but with additional features to handle geospatial data.

### Display data frame entries

- `CVEGEO`: The CVEGEO is a unique geo-statistical code assigned by INEGI. This code is used to uniquely identify geographic areas for statistical purposes, allowing for accurate data collection and analysis across different regions in Mexico.
- `CVE_ENT`: The CVE_ENT is the state code provided by INEGI. This code is a numeric identifier for each of the states within Mexico, facilitating the organization and retrieval of state-level data in geographic and statistical analyses.
- `CVE_MUN`: The CVE_MUN is the municipality code provided by INEGI. Each municipality within a state is assigned a unique CVE_MUN code, which is used to distinguish and manage data specific to municipalities within the broader context of state and national statistics.
- `NOMGEO`: The NOMGEO represents the municipality name as provided by INEGI. This is the official name of the municipality, used in conjunction with its CVE_MUN code to provide a human-readable identifier for geographic and statistical records.
- `geometry`: The geometry attribute contains the boundary information of each municipality as defined by INEGI. This is typically represented as polygons, which outline the geographic area of the municipality. These polygons are crucial for spatial analyses, mapping, and visualizations, as they define the physical extents of each municipality.

### Mexico has 32 states.  
Mexico, officially known as the United Mexican States, is a country located in the southern part of North America. It is composed of 32 states, each with its unique cultural, geographical, and historical attributes. Mexico covers an area of approximately 1.96 million square kilometers, making it the 14th largest country in the world.

The capital city, Mexico City (Ciudad de México), is one of the most populous and important cultural and economic centers globally. Founded by the Aztecs as Tenochtitlán in 1325, it was later rebuilt by the Spanish and has since grown into a sprawling metropolis. Mexico City is known for its rich history, vibrant cultural scene, and significant political and financial influence.

Geographically, Mexico is diverse, encompassing deserts, mountains, forests, and coastal plains. It has extensive coastlines along the Pacific Ocean, the Gulf of Mexico, and the Caribbean Sea, contributing to its varied climate and ecosystems. The country is home to several mountain ranges, including the Sierra Madre Occidental and the Sierra Madre Oriental, and prominent volcanoes like Popocatépetl and Iztaccíhuatl.

Demographically, Mexico has a population of over 126 million people, making it the 10th most populous country in the world. The population is a blend of indigenous peoples, mestizos (mixed indigenous and European ancestry), and people of European descent. Mexico has a rich cultural heritage, with numerous indigenous languages and traditions preserved alongside Spanish, which is the official language.

Historically, Mexico has a deep and complex history that includes the rise and fall of ancient civilizations such as the Maya and Aztec, colonization by Spain in the 16th century, and a long struggle for independence, which was achieved in 1821. The country's history is reflected in its many archaeological sites, colonial cities, and cultural practices.

Economically, Mexico has a mixed economy with significant contributions from agriculture, manufacturing, and services. It is one of the world's largest producers of silver, oil, and automobiles. Tourism is also a vital part of the economy, with millions of visitors attracted to its historic sites, beautiful beaches, and vibrant culture.

```python
# Plot Mexico's map labeling each state by a color. 
# States are identified bya a CVE_ENT code

ax = gdf.plot(figsize = (60, 60),column="CVE_ENT",cmap="Greens")
ax.set_axis_off()
```

- `ax = gdf.plot(figsize = (60, 60), column="CVE_ENT", cmap="Greens")`: This line creates a plot of the GeoDataFrame gdf. The figsize = (60, 60) parameter sets the size of the plot. The column="CVE_ENT" parameter specifies that the states should be colored according to their CVE_ENT code. The cmap="Greens" parameter sets the color map to "Greens".
- `ax.set_axis_off()`: This line removes the axis from the plot for a cleaner visual representation.

### Each Mexican state is subdivided into municipalities  
Each Mexican state is subdivided into municipalities, which are the smallest administrative units in the country. Municipalities, known as "municipios" in Spanish, function similarly to counties in the United States. They are governed by a municipal president (mayor) and a municipal council, both of which are elected by the residents. The municipal president oversees the local government and administration, while the council is responsible for legislative functions.

Municipalities vary widely in terms of size, population, and economic activity. Some encompass large urban areas with millions of residents, while others consist of small rural communities with fewer inhabitants. For instance, Mexico City, although it is a federal entity, is divided into 16 boroughs (delegaciones) that function like municipalities.

The structure and number of municipalities within each state can differ significantly. For example, the state of Oaxaca has the largest number of municipalities, totaling 570, many of which are small and indigenous. In contrast, states like Baja California have fewer municipalities, reflecting different administrative and geographic considerations.

Municipalities are pivotal in the federal structure of Mexico, as they ensure that local governance is accessible and responsive to the citizens. They also provide a means for local communities to participate in the democratic process and have a say in how their local affairs are managed.
