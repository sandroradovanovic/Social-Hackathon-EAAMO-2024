### Load libraries

- `import geopandas as geopd`: This line imports the geopandas library and assigns it the alias geopd. GeoPandas is a library that facilitates working with geospatial data in Python, extending pandas capabilities to include spatial operations on geometric types.
- `import pandas as pd`: This line imports the pandas library and assigns it the alias `pd`. Pandas is a data manipulation and analysis library that provides flexible and powerful data structures like DataFrames.

### Load San Luis Potosi state shape file  
Source: Marco Geoestadístico INEGI (2023).  

#### URL: https://www.inegi.org.mx/temas/mg/ 

```python
# INEGI data
gdf = geopd.read_file("../dataset/San Luis Potosi/geometry/2023_1_24_MUN/2023_1_24_MUN.shp")
```

- INEGI data: The following code is dealing with data from INEGI (Instituto Nacional de Estadística y Geografía), Mexico's National Institute of Statistics and Geography.
- `gdf = geopd.read_file("../dataset/San Luis Potosi/geometry/2023_1_24_MUN/2023_1_24_MUN.shp")`: This line reads a shapefile using GeoPandas. The file path points to the shapefile on the disk, containing geographic data for the state of San Luis Potosi. The read file is stored in the variable gdf as a GeoDataFrame, a data structure that supports spatial operations.

### Display data frame entries

- `CVE_ENT`: The CVE_ENT is the state code provided by INEGI. This code is a numeric identifier for each of the states within Mexico, facilitating the organization and retrieval of state-level data in geographic and statistical analyses.
- `CVE_MUN`: The CVE_MUN is the municipality code provided by INEGI. Each municipality within a state is assigned a unique CVE_MUN code, which is used to distinguish and manage data specific to municipalities within the broader context of state and national statistics.
- `NOMGEO`: The NOMGEO represents the municipality name as provided by INEGI. This is the official name of the municipality, used in conjunction with its CVE_MUN code to provide a human-readable identifier for geographic and statistical records.
- `geometry`: The geometry attribute contains the boundary information of each municipality as defined by INEGI. This is typically represented as polygons, which outline the geographic area of the municipality. These polygons are crucial for spatial analyses, mapping, and visualizations, as they define the physical extents of each municipality.

- `display(gdf)`: This line displays the GeoDataFrame gdf. In a Jupyter notebook, this will render the GeoDataFrame as a table, showing its structure and contents, including columns like CVE_MUN and geometry.

### San Luis Potosi has 58 municipalities  
San Luis Potosí is a state located in the central region of Mexico. It is known for its diverse geography, which ranges from arid deserts to lush forests and includes significant portions of the Sierra Madre Oriental mountain range. The state covers an area of approximately 61,137 square kilometers and is bordered by nine other Mexican states.

The capital city, also named San Luis Potosí, serves as a major economic and cultural hub for the region. Founded in 1592, the city has a rich history influenced by mining activities, particularly silver mining, which spurred its growth during the colonial period. The city's historic center, with its well-preserved colonial architecture, has been recognized as a UNESCO World Heritage site.

Demographically, San Luis Potosí is home to a population of over 2.8 million people, according to the latest census. The population is a mix of indigenous communities, mestizos, and people of European descent. The state is known for its cultural diversity, with various indigenous groups, including the Huastec and Pame peoples, contributing to the region's cultural heritage.

The economy of San Luis Potosí is diverse, encompassing agriculture, manufacturing, mining, and services. The state is a significant producer of crops such as sugar cane, corn, and beans, and it has a growing industrial sector with numerous manufacturing plants, including automotive and aerospace industries.

San Luis Potosí is also a destination for tourism, offering natural attractions such as the Huasteca Potosina region, known for its waterfalls, rivers, and caves, and historical sites like the El Sótano de las Golondrinas, one of the world's deepest pits, popular among spelunkers and adventure tourists.

```python
# Plot Mexico's map labeling each state by a color. 
# Municipalities are identified bya a CVE_MUN code

ax = gdf.plot(figsize = (60, 60),column="CVE_MUN", cmap = "Greens")
ax.set_axis_off()

# Include municipalities outlines.
gdf.boundary.plot(ax=ax,figsize = (20, 20), edgecolor='midnightblue')
ax.set_axis_off()
```

- `ax = gdf.plot(figsize = (60, 60), column="CVE_MUN", cmap = "Greens")`: This line creates a plot of the GeoDataFrame gdf. The figsize = (60, 60) parameter sets the size of the plot. The column="CVE_MUN" parameter specifies that the municipalities should be colored according to their CVE_MUN code. The cmap = "Greens" parameter sets the color map to "Greens".
- `ax.set_axis_off()`: This line removes the axis from the plot for a cleaner visual representation.
- `gdf.boundary.plot(ax=ax, figsize = (20, 20), edgecolor='midnightblue')`: This line adds the boundaries of the municipalities to the existing plot. The ax=ax parameter specifies that the boundaries should be plotted on the same axes as the previous plot. The figsize = (20, 20) parameter is redundant here because the plot size is already set. The edgecolor='midnightblue' parameter sets the color of the boundaries to midnight blue.
- `ax.set_axis_off()`: This line removes the axis from the plot for a cleaner visual representation.
