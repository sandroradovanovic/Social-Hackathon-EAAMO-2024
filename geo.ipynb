```python
import geopandas
import pandas as pd
```

```python

gdf = geopandas.read_file("2023_1_24_A/2023_1_24_A.shp")

gdf_slp = gdf[gdf["CVE_MUN"] == "028"]
gdf_sgs = gdf[ gdf["CVE_MUN"] == "035"]

frames = [gdf_slp, gdf_sgs ]

gdf_zmslp = pd.concat(frames)
```

```python
import seaborn as sns
from shapely import box





gdf_v = geopandas.read_file("INEGI_DENUE_29062024/INEGI_DENUE_29062024.shp")


gdf_slp.plot()
gdf_sgs.plot()
gdf_zmslp.plot()
base = gdf_zmslp.boundary.plot(figsize = (20, 20), edgecolor='black')
base.set_axis_off()
base.set_title('ZMSLP')

base = gdf_v.plot(ax=base, figsize = (30, 30), color = 'red')


df = pd.read_csv("resumenEspacios_updated2.csv")

gdf_escuelas = geopandas.GeoDataFrame(
    df, geometry=geopandas.points_from_xy(df.longitude, df.latitude), crs="WGS84"
)

ax2 = gdf_escuelas.plot(ax=base, figsize = (30, 30), color = 'green', marker='o', markersize = 100)


ax3 = sns.kdeplot(ax= ax2,data=gdf_v, x="longitud", y="latitud", fill=True,cmap="Reds",cut=3,bw_adjust=.8,levels=15,gridsize=400,cbar=True)

ax4 =gdf_escuelas.plot(ax=ax3, figsize = (30, 30), color = 'blue', marker='o', markersize = 100)

ax5= gdf_v.plot(ax=ax4, figsize = (30, 30), color = 'red')

base = gdf_zmslp.boundary.plot(ax=ax5, figsize = (30, 30), edgecolor='black')


#xMin, yMin, xMax, yMax = gdf_zmslp.total_bounds
#box = geopandas.GeoDataFrame(geometry=[box(xMin, yMin, xMax, yMax)])
#box.plot(ax=base, color="black", zorder=0)
```
